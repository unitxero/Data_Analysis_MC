```python
import numpy as np
import pandas as pd
```

```python
# The data is from 2010
mlb = pd.read_csv('https://raw.githubusercontent.com/UwzH43ZRx-5S/mlData/main/mlb.csv')
```

```python
df_mlb = mlb
```

```python
# preview of the data set
# showing the first 5 and last 5 rows
# there are 4 columns 
df_mlb
```

```python
# we return a result of 828 rows, in a step of 1
df_mlb.index
```

```python
# we return 828 rows and 4 columns
df_mlb.shape
```

```python
# we rename some columns here, and use the inplace to True to 
# keep our work saved
# also multiplied the salary which is in 1000's to millions to display
# the true value

# two column changes and one broadcast

df_mlb.rename(columns={'player': 'Player'}, inplace=True)
df_mlb.rename(columns={'team': 'Team'}, inplace=True)
df_mlb['Actual Salary'] = df_mlb['salary'].mul(1000)

df_mlb
```

```python
# Here we are finding the highest paid player in MLB
# which comes out to Alex Rodriguez

print(df_mlb[['Player', 'salary']][df_mlb.salary == df_mlb['salary'].max()])
# Highest paid player in the MLB at the time was Alex Rodriguez.
# At 33 Million
```

```python
# Here we are finding the highest paid Shortstop
# We use the loc command to get it by column name, and for Shortstop filtering
shortstops = df_mlb.loc[df_mlb['position'] == 'Shortstop']
print(shortstops[['Player', 'salary']][shortstops.salary == shortstops['salary'].max()])
# The highest paid shortstop was Derek Jeter.
# At 22.6 Million
```

```python
# Here we are finding the highest paid third baseman
# We use the loc command to get it by column name, and for Third Baseman filtering
thirdbaseman = df_mlb.loc[df_mlb['position'] == 'Third Baseman']
print(thirdbaseman[['Player', 'salary']][thirdbaseman.salary == thirdbaseman['salary'].max()])
# The highest paid third baseman was Alex Rodriguez.
# At 33 Million
```

```python
# Here we are finding the highest paid catcher
# We use the loc command to get it by column name, and for catcher filtering
catcher = df_mlb.loc[df_mlb['position'] == 'Catcher']
print(catcher[['Player', 'salary']][catcher.salary == catcher['salary'].max()])
# The highest paid catcher was Jorge Posada.
# At 13.1 Million
```

```python
# Here we are finding the highest paid Pitcher
# We use the loc command to get it by column name, and for Pitcher filtering
pitcher = df_mlb.loc[df_mlb['position'] == 'Pitcher']
print(pitcher[['Player', 'salary']][pitcher.salary == pitcher['salary'].max()])
# The highest paid pitcher was CC Sabathia.
# At 24.28 Million
```

```python
# Here we are finding the highest paid Outfielder
# We use the loc command to get it by column name, and for Outfielder filtering
outfielder = df_mlb.loc[df_mlb['position'] == 'Outfielder']
print(outfielder[['Player', 'salary']][outfielder.salary == outfielder['salary'].max()])

# The highest paid outfielder was Carlos Beltran.
# At 19.4 Million
```

```python
# Here we are finding the highest paid player on Arizona Diamondbacks
# We use the loc command to get it by column name, and for Arizona Diamondbacks
# Then we grabbed the player and salary column, and applied the .max() function
# to get the highest paid player

teamy = df_mlb.loc[df_mlb['Team'] == 'Arizona Diamondbacks']
print(teamy[['Player', 'salary']][teamy.salary == teamy['salary'].max()])
#The highest paid player on the Arizona Diamondbacks was Brandon Webb.
```

```python
teamsum = df_mlb[['Team', 'salary']]
```

```python
teamgroups = teamsum.groupby(by= ['Team'])
teamsal=teamgroups.agg({'salary':"sum"})
teamsal.reset_index(inplace=True)
teamsal
# Teamsum gives you a list of players and their salaries, with only the information regarding their team and their salary.
# From there, I created teamgroups in order to tell Python to group the dataset by team for the next process
# Then, I used the '.agg' function to have the salaries within each group to be mutated into a total sum per group.
# Finally, I used reset_index in order to make it an ordered list once again.
```

```python
avgsals=teamgroups.agg({'salary':"mean"})
avgsals.reset_index(inplace=True)
avgsals
```

```python


print(teamsal[['Team', 'salary']][teamsal.salary == teamsal['salary'].max()])
# The team with the highest salary expense overall was the New York Yankees.
# which was 206.33 Million
```

```python
print(avgsals[['Team', 'salary']][avgsals.salary == avgsals['salary'].max()])
# The team with the highest average pay overall was the New York Yankees.
# which was 8.25 Million
```
