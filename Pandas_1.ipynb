```python
df_nba.head()
```

```python
# the .add fucntion, adds 5 to the values of that column
df_nba['Age'].add(5)
# same as doing df_nba['Age'] + 5 but not something we should use
```

```python
df_nba['Age'] + 5
```

```python
# subtracting 50,000 from the salary column
# same as df_nba['Salary] - 50000

df_nba['Salary'].sub(50000)
```

```python
df_nba.head(3)
```

```python
# change weight to kg 
# so lets convert this by multiplying the weight
df_nba['Weight'].mul(0.453592)
```

```python
df_nba['Weight_in_kgs'] = df_nba['Weight'].mul(0.453592)
```

```python
df_nba
```

```python
# create a salary in millions
# taking the salary column, divide by 1 million
# taking the round command to cut off at 2 decimal places
# then saving it as a new column

df_nba['Salary_in_millions'] = df_nba['Salary'].div(1000000).round(2)
```
