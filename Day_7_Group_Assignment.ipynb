```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
```

```python
# checking to see that df_election remains untouched and usable as source variable
# for dataset
df_election.head()
```

```python
# verifying the data types in the lastest graph prep variable
df_graph_prep_3.info()
```

```python
# creating a new variable that holds the 'Total Ballots Counted (Estimate)'
# this will then be cleaned by removing the ',' at a later time
df_ballot_count = df_election['Total Ballots Counted (Estimate)']
df_ballot_count.head()
```

```python
# creating a new graph_prep variable to leave the previous unchanged as 
# this is new section for data analysis

df_graph_prep_4 = df_graph_prep_3
```

```python
# adding a column to graph prep 4
# from the previously created variable that contained
# the ballot count columns
df_graph_prep_4['Total Ballot Count Estimate'] = df_ballot_count
```

```python
# verifying that they merged properly
# note that the new column is still and object type
df_graph_prep_4.info()
```

```python
# remove commas from column in ballot count in previous variable 
df_ballot_count.replace(',', '', regex = True, inplace= True)
```

```python
# saving the type change into a var called col_insert, so we can insert it into graph_prep 4

col_insert = df_ballot_count.astype({"Total Ballots Counted (Estimate)": int}, errors = 'raise')
```

```python
# adding a new column to graph_prep_4 from col_insert
df_graph_prep_4['Total Ballot Count (Estimated)'] = col_insert
```

```python
df_graph_prep_4.info()
```

```python
df_graph_prep_4.drop(labels = 'Total Ballot Count Estimate', axis = 1, inplace = True)
```

```python
df_state_prep = df_election['State Abv']
df_state_prep.head()
```

```python
df_state_prep = df_state_prep.dropna()
```

```python
# Final data 
df_graph_final = df_graph_prep_4
```

```python
df_graph_final.drop(labels = 'State', axis = 1, inplace = True)
```

```python
df_graph_final['State'] = df_state_prep
```

```python
df_graph_final
```

```python
# sorting values so the chart is more readable from left to right
df_graph_final = df_graph_final.sort_values(by = 'VEP1',ascending= False)
```

```python
# creating a subplot, and assigning the figure size
fig, ax = plt.subplots(figsize=(30,12))

# creating stacked bar charts
# the VEP1 bar chart will be marked in grey to show the total voting eligible people
# the red bar is the second bar chart to overlap the grey bar to show the
# disparty between the ballots cast
ax = sns.barplot(x = 'State', y = 'VEP1', data=df_graph_final, color ='grey')
ax = sns.barplot(x = 'State', y = 'Total Ballot Count (Estimated)', data=df_graph_final, color = 'red')

# testing to see if a legend would appear
# more testing is needed to allocate its color
# and size
plt.legend(labels = ["VEP", "Total Ballot Count (Estimated)"], title = 'VEP vs Total Ballot Count (Estimated)')
plt.show()
```

## Insight 5

In this graph we are looking at voting-eligible population and
total ballots cast.
We see an interesting trend of the top 5 states once again appearing.

California, Texas, Florida, New York, Pennsylvania

There is a huge difference in ballots cast (estimated) to the amount of people who can actually vote in California. 9.6 Million people estimated, who were eligible did not vote.
In a previous insight, we saw that California had the most VEP meaning they could swing elections with their 55 electoral votes.

Political canidates have 9.6 Million missing votes they can potentially win to swing the state in their favor. Those 55 electoral votes can be pivotal in a campaign win. Texas holds very similar possiblities with the amount of VEP that do not cast votes.


```python
# setting the width variable with anchos to 0.5

anchos = 0.5

# passing the .Figure function from plotly 

fig = go.Figure()

# assigning the x-axis and y-axis for 'Voting Eligible Population'
# this will be the graph in the 'back' showing the total eligible population
# in blue
fig.add_trace(go.Bar( x = df_graph_final['State'],
                      y = df_graph_final['VEP1'],
                     width = anchos, name = 'Voting Eligible Population'
                     ))

# assigning the x-axis and the y-axis for 'Total Ballot Count (Estimated)'
# this will be the graph in the 'front' showing the estimated ballot count
# an offset is introduced here, to make it more apparent
fig.add_trace(go.Bar(x = df_graph_final['State'],
                     y = df_graph_final['Total Ballot Count (Estimated)'],
                     offset = 0.0005,
                     width = anchos,
                     name = 'Ballot Count Estimated'
                     ))

# the title is named here, and the size of the chart is set to 1280x1024
# font size is set to 30
fig.update_layout(title = "Voting Eligible Population to Ballots Cast (Estimated) per State",
               barmode = "overlay",
               title_font_size = 30,
               width = 1280,
               height = 1024)
# X axis is updated to State
fig.update_xaxes(
                title_text = 'State'
)
# Y axis is updated to 'Voter Eligible Population to Ballots Cast'
fig.update_yaxes(
                title_text = 'Voter Eligible Population to Ballots Cast'
)
# How does the eligible population compare to the ballot count of each state?
# Thus, we can see which state had the greatest turnout rate as well. 
```
